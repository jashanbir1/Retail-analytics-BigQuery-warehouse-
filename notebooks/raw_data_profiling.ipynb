Final profiling conclusions:
	•	Raw grain is one invoice line item per product.
	•	InvoiceDate is initially stored as text and should be converted to a proper datetime type.
	•	CustomerID contains many nulls and cannot be assumed to be present for every transaction row.
	•	Description contains nulls, and all null-description rows also have null CustomerID and UnitPrice == 0, indicating a specific low-trust subgroup 	 rather than random missingness.
	•	Cancellation rows are identified by InvoiceNo values beginning with C.
	•	Negative-quantity rows overlap strongly with cancellation rows, but they are not exactly the same population, so they should be treated as related but separate business conditions.
	•	There are a small number of negative-price rows that appear to be accounting or adjustment-style records rather than normal customer sales.
	•	There are many zero-price rows, and a large portion of them also have null CustomerID; these rows likely include a mix of incomplete, promotional, free-item, and other nonstandard operational records.
	•	The combination of UnitPrice == 0 + null Description + null CustomerID forms a distinct suspicious subgroup that should not be treated as normal retail sales.
	•	Exact duplicate rows exist across all columns and should be preserved in bronze but deduplicated in silver while keeping one copy.
	•	The source does not appear to contain a separate invoice-line surrogate key, so invoice-line grain must be inferred from the available business columns.
	•	The dataset supports deriving customer, product, country, and date dimensions, with a sales fact table at invoice-line grain.
